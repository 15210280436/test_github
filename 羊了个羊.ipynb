## 羊了个羊逻辑拆解
### 卡牌位置（X,Y,Z）=14✖️3✖️6=252个
#### 14种图案，3个相同才能消除，大概有6组
### 假设前面顺利消除，残局阶段


```python
# 残局阶段14张卡片+底部最多有6张卡片有N种图案才能消除，那么卡片至少是3N
# 那么核心条件：3N-14<=6才可能消除，得到N<=6种图案
# 模拟
# 1、生成卡片数量14*3*6=252（生成后打乱）
# 2、假设前14张卡片是残局，统计图案数量N
# 3、N<=6 记录

import random
import pandas as pd

# 记录252张卡片
cards=[]

# 记录N，查看分布
counter=[]

df_temp=pd.DataFrame(dtype='int')
# 生成卡片数量
for i in range(15):
    # 直接统计i就可以表示图案
    for j in range(18):
        cards.append(i)

# 模拟1000次查看分布
for t in range(1000):
    # 计数
    N=0
    # 模拟10000次
    for i in range(10000):
        # 洗牌
        random.shuffle(cards)
        # 统计前14张牌图案种类个数
        n=len(set(cards[0:14]))
        # 如果图案数量小于7则计数
        if n<7:
            N=N+1
    counter.append(N)
counter
        
```
